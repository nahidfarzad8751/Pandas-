## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
age_data = df[["Gender" , "Age" , "SN"]]

age_data
```

```python
drop_dupes = age_data.drop_duplicates()

drop_dupes
```

```python



bins = [0, 9.9, 14.9, 19.9, 24.9, 29.9, 34.9, 49.9, 1000]

labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']
    
    
    
drop_dupes["Age Range"] = pd.cut(drop_dupes["Age"], bins ,labels = labels)

grouped_age_df = drop_dupes.groupby(["Age Range"])


#count 

count = grouped_age_df["Age"].count()
```

```python
#percentages 

percentages = (count/total_players)*100


#dataframe display
age_demographics = pd.DataFrame({"Percentages":percentages,
                                 "Counts":count})


age_demographics 
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_ID_df = df[[ "Age" , "Price", "SN", "Purchase ID"]]

drop_dupes_age = purchase_ID_df.drop_duplicates()

drop_dupes_age
```

```python

drop_dupes_age["Age Range"] = pd.cut(drop_dupes_age["Age"], bins ,labels = labels)


age_purchase_count = drop_dupes_age["Purchase ID"].count()

age_average_purchase_price = drop_dupes_age["Price"].mean()

age_total_revenue = drop_dupes_age["Price"].sum()

average_purchase_price_pp = age_total_revenue /100




# dataframe 
age_analysis = pd.DataFrame ({"Purchase count": [age_purchase_count] , 
                              "Average purchase price": [age_average_purchase_price] ,
                              "Total Purchase Value":[age_total_revenue] ,
                              "Average total purchase price per person":[average_purchase_price_pp] })


age_analysis = age_analysis.reindex(['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+'])




age_analysis

#I dont know why the number are not coming up!
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python


top_spenders = df[["Item Name" , "Purchase ID" , "Price" , "SN"]]


top_spenders = top_spenders.groupby(["SN"])

```

```python
purchase_count = top_spenders["Purchase ID"].count()

average_purchase_price = top_spenders["Price"].mean()

total_purchase_value = top_spenders["Price"].sum()
```

```python
purchase_count 
```

```python
#dataframe
spenders = pd.DataFrame({"Purchase count": purchase_count ,
                             "Average purchase price": average_purchase_price ,
                             "Total purchase value":  total_purchase_value })



#pull the top five biggest spenders
formatted_spenders = spenders.sort_values(["Total purchase value"], ascending = False).head()


formatted_spenders
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular_items = df[["Item ID","Item Name", "Price"]]

most_popular_grouped = most_popular_items.groupby(["Item ID", "Item Name" ])

```

```python

purchase_count = most_popular_grouped["Price"].count()
total_purchase_value = most_popular_grouped["Price"].sum()


popular = pd.DataFrame({"Purchase count":purchase_count,
                        "Total purchase value":total_purchase_value })



formatted_popular_items = popular.sort_values(by=['Purchase count'], ascending=False)

formatted_popular_items.head()
```
